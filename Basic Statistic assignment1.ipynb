```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file into a DataFrame
Sales_df = pd.read_csv('sales_data_with_discounts.csv')

# Set the style of seaborn
sns.set(style="whitegrid")

```

```python
# Histograms
num_columns = Sales_df.select_dtypes(include=['float64', 'int64']).columns

for col in num_columns:
    plt.figure(figsize=(5, 6))
    sns.histplot(Sales_df[col], kde=True, bins=30)
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
# Boxplots
for col in num_columns:
    plt.figure(figsize=(10, 6))
    sns.boxplot(y=Sales_df[col])
    plt.title(f'Boxplot of {col}')
    plt.ylabel(col)
    plt.show()
```

```python
# Identify categorical columns
cat_columns = Sales_df.select_dtypes(include=['object', 'category']).columns

# Set the style of seaborn
sns.set(style="whitegrid")

# Bar Chart Analysis for Categorical Columns
for col in cat_columns:
    plt.figure(figsize=(10, 6))
    sns.countplot(y=Sales_df[col], order=Sales_df[col].value_counts().index)
    palette = 'winter'
    plt.title(f'Bar Chart of {col}')
    plt.xlabel('Count')
    plt.ylabel(col)
    plt.show()
```

```python
# Identify numerical columns
num_columns = Sales_df.select_dtypes(include=['float64', 'int64']).columns

# Function to standardize a column
def standardize(column):
    return (column - column.mean()) / column.std()

# Plot original distributions
plt.figure(figsize=(25, 20))
for i, col in enumerate(num_columns, 1):
    plt.subplot(len(num_columns), 2, i)
    sns.histplot(Sales_df[col], kde=True, bins=30)
    plt.title(f'Original Distribution of {col}')
plt.tight_layout()
plt.show()

# Standardize numerical columns
Sales_df_standardized = Sales_df.copy()
for col in num_columns:
    Sales_df_standardized[col] = standardize(Sales_df[col])

# Plot standardized distributions
plt.figure(figsize=(25, 20))
for i, col in enumerate(num_columns, 1):
    plt.subplot(len(num_columns), 2, i)
    sns.histplot(Sales_df_standardized[col], kde=True, bins=30)
    plt.title(f'Standardized Distribution of {col}')
plt.tight_layout()
plt.show()
```

```python
# Identify categorical columns
cat_columns = Sales_df.select_dtypes(include=['object', 'category']).columns

# Apply one-hot encoding
Sales_df_encoded = pd.get_dummies(Sales_df, columns=cat_columns)

# Display the first few rows of the transformed dataset
print(Sales_df_encoded.head())
```
